```python
def load_data(data_dir, name, data_type, num_shots):
    data_path = os.path.join(data_dir, str(name), str(data_type)+'_cat_acc_{}shots.json'.format(num_shots))
    with open(data_path, 'r') as f:
        data = json.load(f)
        
    df = pd.DataFrame.from_dict(data, orient='index', columns=['mean', 'std'])
    
    return df
    
```

```python
num_shots = 2
```

```python
r_test = load_data('experiments/', 'reptile', 'test', num_shots)
# r_train = load_data('experiments/', 'reptile', 'train', num_shots)
j_test = load_data('experiments/', 'joint', 'test', num_shots)
# j_train = load_data('experiments/', 'joint', 'train', num_shots)
```

```python
test_comp = r_test.join(j_test, lsuffix='_reptile', rsuffix='_joint')
# test_comp.head(20)
```

```python
test_comp.loc[test_comp['mean_reptile'] > test_comp['mean_joint'], 'r_beat_j'] = 1
```

```python
def f(row):
    if row['mean_reptile'] > row['mean_joint']:
        val = '1'
    elif row['mean_reptile'] < row['mean_joint']:
        val = '-1'
    else:
        val = '0'
    return val

test_comp['r_beat_j'] = test_comp.apply(f, axis=1)
test_comp['r_beat_j'].describe()
```

```python
with open('data/preprocess_data/splitted_data_pos_6.json', 'r') as f:
    raw_dataset = json.load(f)
cat_companies_id_dict = raw_dataset['test']

len(cat_companies_id_dict['Home & Garden'])
```

```python
r_win_test = test_comp[test_comp['r_beat_j'] == '1']
r_win_test['delta_mean'] = r_win_test['mean_reptile'] - r_win_test['mean_joint']
# r_win_test.sort_values(by=['delta_mean'])

r_win_test['num_companies'] = [len(cat_companies_id_dict[cat]) for cat in r_win_test.index]
r_win_test.sort_values(by=['delta_mean'])
```
